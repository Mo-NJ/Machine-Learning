# 1️⃣ Data Wrangling (Data Preparation)
## Goal: Make raw data clean, usable, and structured.

Data wrangling is about taking messy, incomplete, inconsistent data and turning it into a clean dataset ready for analysis or modeling.

### Typical Tasks:
- Handling missing values
- Removing duplicates
- Fixing inconsistent formats (dates, currencies, casing)
- Standardizing units
- Merging datasets
- Filtering irrelevant records
- Correcting data types

## Example:
You receive a CSV where:
- Dates are in multiple formats
- Some rows have missing emails
- State names are sometimes abbreviated, sometimes spelled out

## Think of it as:
1) <font color=blue>🧹 Making the data correct and usable.</font>
2) <font color=blue>Cleaning and standardizing all of that = data wrangling.</font>


# 2️⃣ Feature Engineering
## Goal: Create better inputs (features) for machine learning models.

Feature engineering happens after the data is clean. It’s about transforming data to improve model performance.

## Typical Tasks:
- Creating new variables (e.g., age from birthdate)
- Encoding categorical variables (one-hot encoding)
- Scaling or normalizing values
- Creating interaction features
- Extracting time-based features (day of week, month, season)
- Aggregating behavior metrics (e.g., total purchases last 30 days)

## Example:
From:
- purchase_history
You create:
- avg_purchase_value
- days_since_last_purchase
- customer_lifetime_value

That’s feature engineering.

## Think of it as:
<font color=blue> 🧠 Making the data more predictive.</font>
